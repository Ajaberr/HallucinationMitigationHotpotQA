```python
import json

def load_jsonl_native(filepath):
    data = []
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line.strip()))
    return data

loaded_data = load_jsonl_native('hotpot_reasoning_10.jsonl')
print(loaded_data)
```

```python
for ex in loaded_data:
    print(f"Question: {ex["question"]}")
    print(f"Reasoning: {ex["reasoning_trace"]}")
    print(f"Answer: {ex["answer"]}")

    model_answer = ex["reasoning_trace"][ex["reasoning_trace"].rfind("the answer is") + len("the answer is "):-1]

    print(model_answer)
    
    print("\n")
```
